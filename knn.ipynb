```python
import numpy as np
import pandas as pd

from sklearn.model_selection import KFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import make_scorer, accuracy_score, recall_score, precision_score, f1_score

import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(pd.read_csv("acoustic_Features.csv"))
X = df.drop(["Class"], axis = 1)
y = df["Class"]
```

```python
accuracy_list = []
n_accuracy_list = []

f1_list = []
n_f1_list = []

recall_list = []
n_recall_list = []

precision_list = []
n_precision_list = []


kfold = KFold(n_splits=10, shuffle=True)
for k in range(1,31):
    df_model = KNeighborsClassifier(n_neighbors=k, metric="minkowski")
    for i in range(10):
        accuracy_list.append(cross_val_score(df_model, X, y, cv = kfold, scoring=(make_scorer(accuracy_score))).mean())
        f1_list.append(cross_val_score(df_model, X, y, cv = kfold, scoring=(make_scorer(f1_score, average = "weighted"))).mean())
        recall_list.append(cross_val_score(df_model, X, y, cv = kfold, scoring=(make_scorer(recall_score, average = "weighted"))).mean())
        precision_list.append(cross_val_score(df_model, X, y, cv = kfold, scoring=(make_scorer(precision_score, average = "weighted"))).mean())
    
    avg_accuracy = 0
    total_accuracy = 0

    for j in accuracy_list:
        total_accuracy += j
    
    avg_accuracy = total_accuracy / 10
    n_accuracy_list.append(avg_accuracy)
    accuracy_list.clear()

    
    
    avg_f1 = 0
    total_f1 = 0

    for j in f1_list:
        total_f1 += j

    avg_f1 = total_f1 / 10
    n_f1_list.append(avg_f1)
    f1_list.clear()

    
    
    avg_precision = 0
    total_precision = 0

    for j in precision_list:
        total_precision += j
    
    avg_precision = total_precision / 10
    n_precision_list.append(avg_precision)
    precision_list.clear()



    avg_recall = 0
    total_recall = 0

    for j in recall_list:
        total_recall += j
    
    avg_recall = total_recall / 10
    n_recall_list.append(avg_recall)
    recall_list.clear()




    
    
  
        


```

```python
f1 = plt.figure()
f1.set_figwidth(16)
f1.set_figheight(8)

plt.plot(range(1,31), n_accuracy_list, "b", label = "Accuracy");
plt.plot(range(1,31), n_f1_list, "c", label = "F1");
plt.plot(range(1,31), n_precision_list, "g", label = "Precision");
plt.plot(range(1,31), n_recall_list, "r", label = "Recall");
plt.title("Metrics for different k-neighbor numbers between 1-30");
plt.xlabel("K-Neighbors");
plt.ylabel("Scores for Minkowski metric");

plt.legend()
```
